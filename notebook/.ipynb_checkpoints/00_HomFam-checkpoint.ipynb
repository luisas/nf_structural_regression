```python
import os
import pandas as pd
import numpy as np
homfam_path = "/home/luisasantus/Desktop/crg_cluster/data/structural_regression/homFam/"
```

```python
f  = "/home/luisasantus/Desktop/crg_cluster/data/structural_regression/homFam/phoslip_ref.vie"
```

```python
def get_stats(f):
    ls = []
    file = open(f, 'r')
    lines = file.readlines()
    n_sequences = 0 
    for index, line in enumerate(lines):
        #print("Line {}: {}".format(index, line.strip()))
        if ">" in str(line): 
            n_sequences = n_sequences + 1
        else: 
            l = len(line.replace("-", ""))
            ls.append(l)

    entry = pd.DataFrame({"file": [os.path.basename(f).replace(".vie", "")],
                          "n_sequences": [n_sequences],
                          "mean_length": [np.mean(ls)],
                      "min_length": [np.min(ls)],
                      "max_length": [np.max(ls)]})
    return(entry)
    
```

```python
#get_stats("/home/luisasantus/Desktop/crg_cluster/data/structural_regression/homFam/phoslip_ref.vie")
```

```python
summary_homfam = pd.DataFrame()
max_steps = 10
i = 0 
for filename in os.listdir(homfam_path):
    i = i+1
    if(i>max_steps): 
        break
    f = os.path.join(homfam_path, filename)
    # checking if it is a file
    if os.path.isfile(f):
        entry = get_stats(f)
        summary_homfam = summary_homfam.append(entry, ignore_index= True)
```

```python
summary_homfam["family"] = summary_homfam.file.str.split("_",expand=True,)[0]
summary_homfam["subset"] = summary_homfam.file.str.split("_",expand=True,)[1]
summary_homfam.sort_values(by=['family'])

column_names = ["family", "subset", "n_sequences","mean_length", "min_length", "max_length", "file"]
summary_homfam = summary_homfam.reindex(columns=column_names)
```
